```python
import os

# Function to load the reference genome from a FASTA file into a dictionary
def load_genome(fasta_file):
    genome = {}
    with open(fasta_file, 'r') as f:
        chrom = ''
        seq = []
        for line in f:
            if line.startswith('>'):
                if chrom != '':
                    genome[chrom] = ''.join(seq)
                chrom = line[1:].strip()
                seq = []
            else:
                seq.append(line.strip())
        if chrom != '':
            genome[chrom] = ''.join(seq)
    return genome

# Function to get the sequence context (-1 and +1 bp around the position)
def get_ms_gr(genome, chr, pos):
    # Fetch the -1 and +1 positions around the mutation site
    seq = genome[chr][pos-2:pos+1]  # -1 and +1 (1-based indexing)
    return seq


# Example of fetching the +1 and -1 nucleotides for the mutation at chromosome 2, position 25228451
with open('401428.dat', 'r') as file:
    next(file)  # Skip the first line
    for line in file:
        columns = line.split()  # Split the line by whitespace (or use another delimiter)
        print(columns[3], columns[4])  # Columns 4 and 5 correspond to indices 3 and 4
        chr_num = "chr" + columns[3]
        position = int(columns[4])

        sequence_context = get_ms_gr(genome, chr_num, position)
        print(f"Sequence context around position {position} on chromosome {chr_num}: {sequence_context}")
print(f"{chr_num} {position} {sequence_context}")
#chr_num = 'chr19'  # Use 'chr' prefix as used in FASTA headers
#position = 42774946

```
